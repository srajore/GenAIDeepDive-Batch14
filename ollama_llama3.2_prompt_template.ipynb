```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import PromptTemplate
```

```python
prompt = PromptTemplate(
        input_variables=["person"], 
        template="Tell me the key achivements of {person} in four bulleted points, if you don't have information please say I don't know ,please help me with correct responses only"
    )
```

```python
llm = ChatOllama(model="llama3.2:latest")
```

```python
chain = prompt | llm # Seq chain
```

```python
response = chain.invoke({"person": "Virat kohli"})
```

```python
print(response.content)
```
